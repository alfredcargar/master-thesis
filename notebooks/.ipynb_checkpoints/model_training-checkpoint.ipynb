# Linear Regression <a id="linear-regression"></a>

Because the model parameters have a relatively low correlation with the target (aprox. 40%), it suggests a weak linear correlation. The linear regression model is thus expected to perform the worst.

### Train

```python
# split into train and test
x_train, x_test, y_train, y_test = train_test_split(df_features, df_target, test_size=0.2, random_state=seed)
```

```python
scaler = StandardScaler()
algorithm = LinearRegression()

model_lr = Pipeline([
    ('scaler', scaler),
    ('algorithm', algorithm)
])
model_lr.fit(x_train, y_train)
```

```python
y_pred = model_lr.predict(x_test)
```

### Evaluation

```python
print(f"R^2 train: {model_lr.score(x_train,y_train)}")
print(f"R^2 test: {model_lr.score(x_test,y_test)}")
```

As seen from the R^2, the model is capturing some of the target's variance but definetely not a large proportion. The similar results in train and test suggest there is no overfitting, however the low values could possibly represent underfitting, there is certainly room for improvement.

**MAE**, **RMSE**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```

On average the predictions with the linear model differ from the actual value by 3.86 units. The RMSE is more sensitive to large errors then MAE, so a value of 4.66 after taking into account the MAE suggests that there are larger errors, but they aren't extremely large.

The MAE and RMSE suggest the error of the model tipically deviate from the actual values by 4-5 units

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(model_lr, x_train, y_train, cv=5, scoring='neg_mean_squared_error')
cv_rmse = (-cv_scores.mean()) ** 0.5
print(f'Cross-Validated RMSE: {cv_rmse}')
```

### Plots

```python
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
sbn.residplot(x=y_pred, y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 2})
plt.axhline(y=0, color='black', linestyle='--', linewidth=2)
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sbn.scatterplot(x=y_test, y=y_pred, s=60)
sbn.lineplot(x=[y_test.min(), y_test.max()], y=[y_test.min(), y_test.max()], color='red', linestyle='--', lw=2)

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.show()
```

# Tree Regression <a id="tree-regression"></a>

```python
from sklearn.tree import DecisionTreeRegressor
```

In a decision tree model we will need to tune different hyperparameters to achieve optimal results:
1. **max_depth**: the depth of the tree, a greater value means more accuracy but also more computational time.
2. **min_samples_split**: the minimum number of samples required to split a node
3. **criterion**: function used to measure the quality of a split (Gini, entropy, etc.)

### Train

```python
x_train, x_test, y_train, y_test = train_test_split(df_features, df_target, test_size=0.2, random_state=seed)
tree_base = DecisionTreeRegressor(random_state=seed)
```

```python
params = {
    'max_depth': [5,8,12,18,25],
    'min_samples_split': [3, 9, 18, 40],
    'criterion': ["squared_error","poisson","absolute_error"]
}
metrics = ['r2','neg_mean_squared_error']
```

```python
gs = GridSearchCV(
    estimator=tree_base,
    param_grid=params,
    cv=4,
    scoring=metrics,
    refit='r2'
)
gs.fit(x_train,y_train)
```

### Evalution

```python
results = pd.DataFrame(gs.cv_results_)
print(results.head())
```

```python
best_r2 = max(gs.cv_results_['mean_test_r2'])
print("Best cross-validation R² score: ", best_r2)
```
